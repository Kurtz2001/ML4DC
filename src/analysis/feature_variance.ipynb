```python
with h5py.File(dset_path, "r") as f:
    dset = f[dset_name][:]
```

```python
X = dset[:, :2807]
```

```python
variances = np.var(X, axis=0)
```

```python
indices = np.argsort(variances)[::-1]
```

```python
indices_zero = []
parameters_zero = [0] * 7

for f in range(2807):
    indice = indices[f]
    variance = variances[indice]
    
    if variance == 0.0:
        indices_zero.append(indice)
        parameters_zero[indice % 7] += 1
    
    print("%d) variance: %f" % (indice, variance))     
    
```

```python
plt.figure()
plt.bar(np.arange(7), parameters_zero)
plt.xticks(np.arange(7), ("MEAN", "RMS", "Q1", "Q2", "Q3", "Q4", "Q5"))
plt.title("Parameter with zero variance")
plt.ylabel("Count")
plt.show()
```
