```python
import os
import sys
import h5py

from golden_json import GoldenJson
```

```python
# Init variables

# Source directory of original h5 datasets
data_src = "/home/mantydze/data/JetHT2016_original" 

# Destinaton directory for updated h5 datasets
data_dst = "/home/mantydze/data/JetHT2016_multi" 

# Golden JSON location
golden_json_path = "/home/mantydze/ml4dc-off/data/Cert_271036-284044_13TeV_PromptReco_Collisions16_JSON.txt"

gj = GoldenJson(golden_json_path)
```

```python
%%time
for filename in os.listdir(data_src):
    print("SRC: %s/%s" % (data_src, filename))
    
    with h5py.File("%s/%s" % (data_src, filename), "r") as file_src:
    
        # Read dataset
        dset_name = filename.replace(".h5", "")    
        dset = file_src[dset_name][:]    

        # Row count
        dset_rc = dset.shape[0]
        print("Rows: %d" % dset_rc)
        
        for row_index in range(dset_rc):
            run = int(dset[row_index, 2807])
            ls = int(dset[row_index, 2808])
            cls = int(dset[row_index, 2812])
            cls_gj = gj.is_good(run, ls)
            
            dset[row_index, 2812] = cls_gj
            
            if (cls == cls_gj):
                print("%d) run %d | ls %d | cls %d | GJ %d" % (row_index, run, ls, cls, cls_gj))
            
        filename = filename.replace("signal", "good").replace("background", "bad")
        dset_name = filename.replace(".h5", "")    

        with h5py.File("%s/%s" % (data_dst, filename), "w") as file_dst:
            file_dst = file_dst.create_dataset(dset_name, 
                                               chunks=True, 
                                               dtype='float64', 
                                               data=dset[:], 
                                               compression="gzip", 
                                               compression_opts=0)
        print("DST: %s/%s" % (data_dst, filename))
        print("*"*80)
```
